````python
# @title Read image using OpenCV
url = "/content/drive/MyDrive/content/drive/kitten01.jpg"
img = cv.imread(url)
plt.imshow(img)

````

````python

# @title Convert Image Color to RGB
img2 = img
img3 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
plt.imshow(img3)
````

````python

# @title Menampilkan citra grayscale
img_gray = cv.imread("/content/drive/MyDrive/content/drive/kitten01.jpg", cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
````

````python
# @title Menampilkan citra grayscale dengen menambahkan cmap gray
plt.imshow(img_gray, cmap="gray")
````

````python

# @title Menampilkan citra grayscale dengan menambahkan cmap magma
plt.imshow(img_gray, cmap="magma")
````

````python
# @title Melakukan resize citra dalam ukuran 512 x 1024
img4 = cv.resize(cv.cvtColor(img, cv.COLOR_BGR2RGB), (512, 1024))
plt.imshow(img4)
````

````python
# @title Citra RGB ditampilkan dengan ukuran yang lebih besar
img5 = cv.flip(cv.cvtColor(img, cv.COLOR_BGR2RGB), 0)


# Menampilkan canvas dengan ukuran yang lebih besar
fig = plt.figure(figsize=(10, 10)) # Mengubah ukuran menjadi (10, 10)
ax = fig.add_subplot(111)
ax.imshow(img5)
````

````python

# @title Membuat black image sebagai canvas
black_img = np.zeros(shape=(512, 512, 3), dtype=np.int32)
plt.imshow(black_img)
````

````python
# @title Membuat rectangle berdasarkan titik koordinat
cv.rectangle(black_img, pt1=(384, 0), pt2=(510, 150), color=(0, 255, 0), thickness=10)
plt.imshow(black_img)
````

````python

# @title Membuat rectangle dengan koordinat berbeda
cv.rectangle(black_img, pt1=(200, 200), pt2=(300, 300), color=(0, 0, 255), thickness=15)
plt.imshow(black_img)

````

````python

# @title Menambahkan lingkaran pada black image
cv.circle(black_img, center=(100, 100), radius=50, color=(255, 0, 0), thickness=8)
plt.imshow(black_img)
````

````python
# @title Menambahkan garis dari titik (0,0) ke (512, 512)
cv.line(black_img, pt1=(0, 0), pt2=(512, 512), color=(255, 255, 255), thickness=5)
plt.imshow(black_img)
````

````python

# @title Menambahkan text pada black image
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text="Hello", org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img)
````

````python

# @title Membuat black img kembali
black_img2 = np.zeros(shape=(512, 512, 3), dtype=np.int32)
plt.imshow(black_img2)
````

````python

# @title Inisialisasi numpy array dengan tipe data int32
vertices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype=np.int32)
vertices
````

````python

# @title Reshape numpy array
pts = vertices.reshape((-1, 1, 2))
pts
````

````python
# @title Menambahkan polyline pada black image
cv.polylines(black_img2, [pts], isClosed=True, color=(255, 0, 0), thickness=5)
plt.imshow(black_img2)
````

````python

# @title 1. Menggunakan figsize apakah ukuran image pixelnya berubah

# Layout figure 10 x 10
fig_sebelum = plt.figure(figsize=(10, 10))
dpi_sebelum = fig_sebelum.dpi
width_sebelum, height_sebelum = fig_sebelum.get_size_inches()
print("Sebelum perubahan ukuran figure:")
print(f'Height: {int(height_sebelum * dpi_sebelum)}, Width: {int(width_sebelum * dpi_sebelum)}')

# Layout figure 8 x 8
fig_sesudah = plt.figure(figsize=(8, 8))
dpi_sesudah = fig_sesudah.dpi
width_sesudah, height_sesudah = fig_sesudah.get_size_inches()
print("Setelah perubahan ukuran figure:")
print(f'Height: {int(height_sesudah * dpi_sesudah)}, Width: {int(width_sesudah * dpi_sesudah)}')
````

Sebelum perubahan ukuran figure:

Kesimpulan

````python

# @title 2. Menampilkan image dalam channel red-blue dan green-blue saja

# Melakukan extract channel warna
b, g, r = cv.split(img)

# Menggabungkan konversi channel warna red-blue
rb_img = cv.merge((b, np.zeros_like(g), r))

# Menggabungkan konversi channel warna green-blue
gb_img = cv.merge((b, g, np.zeros_like(r)))
````

````python

# Menampilkan gambar red-blue channel
plt.imshow(cv.cvtColor(rb_img, cv.COLOR_BGR2RGB)), plt.title("Channel Red-Blue")
````

````python

# Menampilkan gambar green-blue channel
plt.imshow(cv.cvtColor(gb_img, cv.COLOR_BGR2RGB)), plt.title("Channel Green-Blue")
````

````python

# @title 3. Menampilkan image baris ke 20-115 dan kolom 25-120

# Memotong Gambar dengan baris 20-115 dan kolom 25-120
cropped_img = img[20:115, 25:120]

# Menampilkan cropped image
plt.imshow(cv.cvtColor(cropped_img, cv.COLOR_BGR2RGB)), plt.title("Cropped Image")
````

````python


# @title 4. Menampilkan image baris ke 5-30 dan hanya menggunakan Channel Red saja

# Melakukan pemotongan baris pada image
cropped_img_red = img[5:30, :]

# Melakukan extract channel warna
b, g, r = cv.split(cropped_img_red)

# Melakukan penggabungan warna
r_img = cv.merge((np.zeros_like(b), np.zeros_like(g), r))

plt.imshow(cv.cvtColor(r_img, cv.COLOR_BGR2RGB)), plt.title("Channel Red & Pemotongan baris")
````

````python

# @title 5. Membuat 5 kotak berbagai ukuran dalam 1 image

img_random = cv.imread("/content/drive/MyDrive/content/drive/kitten01.jpg")

width_image = img_random.shape[1]
height_image = img_random.shape[0]

for i in range(5):
  # Melakukan randomisasi starting point
  x1, y1 = random.randint(0, width_image), random.randint(0, height_image)

  # Melakukan randomisasi endpoint
  x2, y2 = random.randint(0, width_image), random.randint(0, height_image)

  # Melakukan randomisasi color rgb
  r, g, b = random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)

  cv.rectangle(img_random, pt1=(x1, y1), pt2=(x2, y2), color=(r, g, b), thickness=3)

# Menampilkan gambar yang sudah ditambah 5 kotak
color_convert_rgb = cv.cvtColor(img_random, cv.COLOR_BGR2RGB)
plt.imshow(color_convert_rgb)
````

````python

# @title 6. Menampilkan Image Dengan Posisi Terbalik

# Membaca image
img_kitten_flip = color_convert_rgb.copy()

# Membalikkan foto
img_flipped = cv.flip(img_kitten_flip, 0)

# Menampilkan image posisi terbalik
plt.imshow(img_flipped)
````

````python
# @title 7. Menampilkan Rectangle pada foto

# Mengambil image foto sendiri dari drive
img_sendiri = cv.imread("/content/drive/MyDrive/content/drive/rafsan.jpg")

# Resize Image
img_sendiri_resized = cv.resize(img_sendiri, (0, 0), fx=(0.1), fy=(0.1))

# Menggambar circle dan rectangle pada wajah
cv.rectangle(img_sendiri_resized, pt1=(210, 120), pt2=(110, 280), color=(0, 0, 255), thickness=3)
cv.circle(img_sendiri_resized, center=(160, 200), radius=50, color=(0, 255, 255), thickness=3)

# Mengubah channel warna & Menampilkan foto
img_edited_colored = cv.cvtColor(img_sendiri_resized, cv.COLOR_BGR2RGB)
plt.imshow(img_edited_colored)
````

````python

# @title 8. Menggambar rectangle pada sudut kiri bawah dari citra kitten

img_kitten = img.copy()

# Menggambar rectangle pada citra kitten
cv.rectangle(img_kitten, pt1=(0, 300), pt2=(200, 200), color=(255, 0, 0), thickness=3)

# Menampilkan gambar rectangle
img_kitten = cv.cvtColor(img_kitten, cv.COLOR_BGR2RGB)
plt.imshow(img_kitten)
````
