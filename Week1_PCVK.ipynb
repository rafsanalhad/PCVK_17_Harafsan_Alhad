```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io, transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
urls = [
    "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
    "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
    "https://th.bing.com/th/id/OIP.2mggqeJ4NZV5yiwdvcq5YwAAAA?rs=1&pid=ImgDetMain"
]

for url in urls:
  image = io.imread(url) # read image
  image = cv.resize(image, (0, 0), fx=0.5, fy=0.5) # resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB) # convert color to rgb
  final_frame = cv.hconcat((image, image_2)) # concat image
  cv2_imshow(final_frame) # showing image
  print('\n')
```

```python
#Langkah 3
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar =", tinggi, "x", lebar)
cv2_imshow(image_2)
```

```python
#Langkah 4
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis horizontal di tengah gambar
for y in range(lebar):
  image_3[int(tinggi/2), y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

Pertanyaan 3:

image = cv.resize(image, (0,0), fx=0.5, fy=0.5)

Kegunaan kode program tersebut adalah untuk melakukan resize/mengubah ukuran gambar menjadi ukuran setengahnya yang dimana fx mengacu pada width dan fy adalah height dari gambar. Jika kita tidak melakukan pengubahan ukuran, gambar yang ditampilkan nanti akan menjadi resolusi/ukuran semulanya.

Pertanyaan 4:


for y in range(lebar):
  image_3[int((tinggi)/2), y] = [255, 255, 255]

Kegunaan dari kode [255, 255, 255] adalah untuk menampilkan warna pada gambar. Format dari matriks warna tersebut merupakan format warna BGR yang dimana jika semua warna memiliki angka 255 maka akan menampilkan warna putih begitupun sebaliknya jika semuanya 0 maka akan menampilkan warna hitam.

Pertanyaan 5:

Keterkaitan antara pixel dan resolusi adalah pixel merupakan 1 kotak matriks warna dari sebuah gambar dan resolusi merupakan seluruh kumpulan kotak/pixel. Jadi jika suatu gambar memiliki jumlah pixel banyak maka resolusi dari gambar tersebut akan semakin jelas/tajam


```python
#Tugas 1
# Membuat garis vertikal pada gambar
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for x in range(tinggi):
  image_3[x, int(lebar/2)] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
# Tugas 2
# Membuat garis diagonal paad gambar
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

cv.line(image_3, (lebar, 0), (0, tinggi), (255, 255, 255), 1)
cv.line(image_3, (0, 0), (lebar, tinggi), (255, 255, 255), 1)

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
# Tugas 3
# Membuat garis dengan panjang tertentu
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

cv.line(image_3, (40, int(tinggi/2)), (100, int(tinggi/2)), [255, 255, 255], 1)

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
#Tugas 4
# Membuat kotak di tengah gambar
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

cv.rectangle(image_3, (int(lebar/2)-20, int(tinggi/2)-20), (int(lebar/2)+20, int(tinggi/2)+20), [255, 255, 255], -1)

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
