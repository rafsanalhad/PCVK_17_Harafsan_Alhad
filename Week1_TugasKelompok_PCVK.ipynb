```python
import matplotlib.pylab as plt
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
from PIL import Image, ImageDraw
```

```python
# Read image with opencv

image = cv.imread('ktp.png')
cv2_imshow(image)
```

```python
# Inisialisasi tinggi dan lebar

tinggi = image.shape[0]
lebar = image.shape[1]
print(f'tinggi: {tinggi}, lebar: {lebar}')
```

```python
image_ktp = cv.cvtColor(image, cv.COLOR_BGR2RGB)

plt.imshow(image_ktp)
plt.axis('off')  # Hilangkan sumbu
plt.title("Sampel")
plt.show()

# Tutup bagian Gol. Darah
image_ktp[200:235, 640:710] = [0, 0, 255] # Warna Biru

# Tutup bagian TTD
image_ktp[500:800, 650:980] = [255, 0, 0] # Warna Merah

# Tampilkan hasil dengan Matplotlib
plt.imshow(image_ktp)
plt.axis('off')  # Hilangkan sumbu
plt.title("Hasil")
plt.show()
```
